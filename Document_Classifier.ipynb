```python
import os
import random
import shutil
import pytesseract
from PIL import Image

# Paths
image_directory = "/content/invoice"
train_text_directory = "/content/contract_invoice/train/invoice"
test_text_directory = "/content/contract_invoice/test/invoice"

# List all image files in the directory
image_files = [file for file in os.listdir(image_directory) if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Choose 302 random image files
random_image_files = random.sample(image_files, 302)

# Perform OCR and store text in train or test text directories
for image_file in random_image_files:
    image_path = os.path.join(image_directory, image_file)
    image = Image.open(image_path)

    ocr_text = pytesseract.image_to_string(image)

    if random.random() < 0.8:  # 80% for training
        text_destination_path = os.path.join(train_text_directory, image_file.replace(".jpg", ".txt"))
    else:
        text_destination_path = os.path.join(test_text_directory, image_file.replace(".jpg", ".txt"))

    # Write OCR text to the destination file
    with open(text_destination_path, "w") as text_file:
        text_file.write(ocr_text)

print("OCR and text storage completed successfully!")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

source_dir = "/content/contract_invoice"  # Replace with your source directory
train_dir = "/content/contract_invoice/train"
test_dir = "/content/contract_invoice/test"

# List all files and their labels
data = []
for label in os.listdir(source_dir):
    if os.path.isdir(os.path.join(source_dir, label)):
        files = os.listdir(os.path.join(source_dir, label))
        files = [os.path.join(label, file) for file in files]
        data.extend([(file, label) for file in files])

# Split data into train and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Create directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Copy files to train directory
for file, label in train_data:
    source_file = os.path.join(source_dir, file)
    dest_file = os.path.join(train_dir, file)
    os.makedirs(os.path.dirname(dest_file), exist_ok=True)
    shutil.copy(source_file, dest_file)

# Copy files to test directory
for file, label in test_data:
    source_file = os.path.join(source_dir, file)
    dest_file = os.path.join(test_dir, file)
    os.makedirs(os.path.dirname(dest_file), exist_ok=True)
    shutil.copy(source_file, dest_file)

print("File structure restored with 80% training and 20% testing.")
```

```python
def build_classifier_model():
  text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
  preprocessing_layer = hub.KerasLayer(tfhub_handle_preprocess, name='preprocessing')
  encoder_inputs = preprocessing_layer(text_input)
  encoder = hub.KerasLayer(tfhub_handle_encoder, trainable=True, name='BERT_encoder')
  outputs = encoder(encoder_inputs)
  net = outputs['pooled_output']
  net = tf.keras.layers.Dropout(0.1)(net)
  net = tf.keras.layers.Dense(1, activation=None, name='classifier')(net)
  return tf.keras.Model(text_input, net)
```
